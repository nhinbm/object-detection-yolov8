### 2. Training

```python
from ultralytics import YOLO
```

#### 2.1. Load configuration files and pre-trained model (yolo8s.pt)

```python
# Load a model
yolo_yaml_path = 'yolo_data/data.yml'
model = YOLO('yolov8s.yaml').load('yolov8s.pt')
```

#### 2.2. Set up some important parameters

```python
# Set up some important parameters
epochs = 15
imgsz = 640
batch_size = 8
patience = 5
lr = 0.0005
```

#### 2.3. Start training (fine-tuning)

```python
results = model.train(
    data=yolo_yaml_path,
    epochs=epochs,
    imgsz=imgsz,
    batch=batch_size,
    lr0=lr,
    patience=patience,
    project='models',
    name='yolov8/detect/train'
)
```

### 3. ƒê√°nh gi√° m√¥ h√¨nh

```python
model_path = 'models/yolov8/detect/train/weights/best.pt'
model = YOLO(model_path)

metrics = model.val(
    project='models',
    name='yolov8/detect/val'
)
```
